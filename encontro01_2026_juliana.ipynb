```python
# programas/bibliotexas utilizados no script
import httpx #responsavel pelas requisições web
from bs4 import BeautifulSoup #responsável por realizar o web scraping (coletar dados)
```

# Fundamentos programação


```python
# variavel e tipo de dados (sting, lista, numero)

#lista

paginas = ["https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=0",
           "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=30",
           "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=60"]
# função: bloco de código que é definida e chamada
def acessar_pagina(link):
  print (f"Estamos na pagina: {link}")
  pag_web = httpx.get(link) #pegando o httpx da pagina (httpx vai na internet )
  bs =  BeautifulSoup (pag_web, "html.parser") #pegando a informação da pagina a partir do httpx
  return bs

#loop
for pagina in paginas:
  acessar_pagina(pagina)
```

```python
# Variável e tipos de dados (string, lista, numero)
paginas = ["https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=0",
          "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=30",
          "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=60",
          "https://www.gov.br/mre/pt-br/canais_atendimento/imprensa/notas-a-imprensa/notas-a-imprensa?b_start:int=90"]

def acessa_pagina (link):
  print (f"Estamos na pagina:{link}")
  pag_web = httpx.get(link, timeout=10.0)
  bs = BeautifulSoup(pag_web, "html.parser")
  return bs

# loop for
# beautifulsoap >> find e find_all
for pagina in paginas:
  pagina_inteira = acessa_pagina(pagina)
  lista_noticias = pagina_inteira.find("div", attrs={"id": "content-core"}).find_all("article")
  for noticia in lista_noticias:
    titulo = noticia.find(attrs={"class": "tileHeadline"}).text.strip()
    print(titulo)
    link_noticia = noticia.a["href"]
    print(link_noticia)
    data = noticia.find(attrs={"class": "summary-view-icon"}).text.strip()
    print(data)
    print("###")
    # numero da nota
    # data
    # horário
    # paragrafos
```
